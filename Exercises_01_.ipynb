**Exercise 3: Loops and Conditionals**

We have a list of view counts: [800, 1200, 3000, 600]

1.   Loop through the list and:

*   Print "Popular!" if views > 1000
*   Otherwise print "Needs more views..."

2.   Count how many articles are "Popular!" in a variable called popular_count.
3.   Print popular_count.


Starter Code

```python
# EXERCISE 3

views_list = [800, 1200, 3000, 600]
popular_count = 0  # We'll count how many articles have views > 1000

# 1. Loop through the list and use an 'if' statement.
#    Print "Popular!" if views > 1000, else "Needs more views..."
# TODO:

# 2. Increment 'popular_count' whenever an article is popular.
# TODO:

# 3. Print 'popular_count'.
# TODO:
```

Solution

```python
# SOLUTION FOR EXERCISE 3

views_list = [800, 1200, 3000, 600]
popular_count = 0

for views in views_list:
    if views > 1000:
        print("Popular!")
        popular_count += 1
    else:
        print("Needs more views...")

print("Number of popular articles:", popular_count)
```

Solution

```python
# SOLUTION FOR EXERCISE 4

import pandas as pd

# Create the data dictionary
data = {
    "Article": ["Climate Report", "Election Update", "Tech Trends"],
    "Views": [1500, 2000, 1200],
    "Likes": [200, 350, 150]
}

# 1. Convert it to a DataFrame called df
df = pd.DataFrame(data)

# 2. Display the first few rows
display(df.head())

# 3. Add the 'Comments' column
df["Comments"] = [50, 120, 30]

# 4. Print the updated DataFrame
print("\nUpdated DataFrame:")
display(df)
```
